```python
import random
import collections

def cont_cor(cor, lista, valorcor):
    x = 0
    resultado = 0
    for i in lista:
        if i == cor:
            resultado = x = x + 1
        cor_cont = resultado > valorcor
    return cor_cont

lista = ['amarelo', 'verde', 'azul', 'roxo', 'vermelho', 'laranja']
azul = 10
amarelo = 10
laranja = 10
roxo = 10
verde = 10
vermelho = 10
x = 10

print('''
Bem-vindo ao Mastermind

Para jogar você deve tentar adivinhar uma senha de 4 cores.
Você deve escolher 4 cores dentre as seguintes:
azul, amarelo, laranja, roxo, verde e vermelho.
Você pode repetir as cores no seu chute, porém não há repetições de cores na senha.
Sempre digite o seu chute separando as cores por vírgula.
Lembre-se que você tem 10 chances.

Para cada cor escolhida você vai receber os seguintes retornos:
Pino Branco: Cor chutada corresponde exatamente a posição dela na senha.
Pino Preto: Cor chutada existe na senha, porém está na posição errada.
Pino Vazio: Cor chutada não existe na senha.

Você ganha o jogo quando todas as cores chutadas estiverem na mesma posição da senha.

''')

senha = random.sample(lista, 4)

while x > 0:
    entrada = input('\nDigite o seu chute:').lower().split(',')
    chute = []
    pos = []
    # Verifica e retira espaço do input
    for i in entrada:
        chute.append(i.strip())
    # Verifica se o input contém 4 itens
    if len(chute) != 4:
        print('Você deve digitar 4 cores')
        continue
    # Verifica se as cores digitadas são válidas
    verifica_cor_diferente = (collections.Counter(chute))
    for i in lista:
        del (verifica_cor_diferente[i])
    if len(list(verifica_cor_diferente)) != 0:
        print('Você deve digitar cores válidas')
        continue
    #Verifica a quantidade de pinos disponíveis para o chute
    if cont_cor('roxo', chute, roxo) == True:
        print(f'Você tem apenas {roxo} pino(s) roxo(s). Tente novamente:')
        continue
    if cont_cor('azul', chute, azul) == True:
        print(f'Você tem apenas {azul} pino(s) azui(s). Tente novamente:')
        continue
    if cont_cor('amarelo', chute, amarelo) == True:
        print(f'Você tem apenas {amarelo} pino(s) amarelo(s). Tente novamente:')
        continue
    if cont_cor('laranja', chute, laranja) == True:
        print(f'Você tem apenas {laranja} pino(s) laranja(s). Tente novamente:')
        continue
    if cont_cor('verde', chute, verde) == True:
        print(f'Você tem apenas {verde} pino(s) verde(s). Tente novamente:')
        continue
    if cont_cor('vermelho', chute, vermelho) == True:
        print(f'Você tem apenas {vermelho} pino(s) vermelho(s). Tente novamente:')
        continue
    # Compara o chute com a senha e retorna os pinos
    else:
        if chute[0] == senha[0]:
            pos.append('branco')
        elif chute[0] == senha[1] or chute[0] == senha[2] or chute[0] == senha[3]:
            pos.append('preto')
        else:
            pos.append('vazio')
        if chute[1] == senha[1]:
            pos.append('branco')
        elif chute[1] == senha[0] or chute[1] == senha[2] or chute[1] == senha[3]:
            pos.append('preto')
        else:
            pos.append('vazio')
        if chute[2] == senha[2]:
            pos.append('branco')
        elif chute[2] == senha[0] or chute[2] == senha[1] or chute[2] == senha[3]:
            pos.append('preto')
        else:
            pos.append('vazio')
        if chute[3] == senha[3]:
            pos.append('branco')
        elif chute[3] == senha[0] or chute[3] == senha[1] or chute[3] == senha[2]:
            pos.append('preto')
        else:
            pos.append('vazio')
        print(pos)
        # Conta a quantidade de chutes disponíveis e retorna para o jogador
        x -= 1
        print('Você tem', x, 'chutes disponíveis')
        # Conta a quantidade de pinos disponíveis e retorna para o jogador
        for qnt_cor in chute:
            if qnt_cor == 'azul':
                azul -= 1
            if qnt_cor == 'laranja':
                laranja -= 1
            if qnt_cor == 'roxo':
                roxo -= 1
            if qnt_cor == 'amarelo':
                amarelo -= 1
            if qnt_cor == 'vermelho':
                vermelho -= 1
            if qnt_cor == 'verde':
                verde -= 1
        print('Você tem', azul, 'pinos azuis disponíveis')
        print('Você tem', laranja, 'pinos laranjas disponíveis')
        print('Você tem', roxo, 'pinos roxos disponíveis')
        print('Você tem', amarelo, 'pinos amarelos disponíveis')
        print('Você tem', vermelho, 'pinos vermelhos disponíveis')
        print('Você tem', verde, 'pinos verdes disponíveis')

        # Compara o chute com a senha e verifica se jogador ganhou
        if chute[0] == senha[0] and chute[1] == senha[1] and chute[2] == senha[2] and chute[3] == senha[3]:
            print('Você ganhou')
            break
if x == 0:
    print('Você esgotou as suas chances')
```
