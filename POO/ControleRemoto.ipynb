```python
class Televisor:
    
    def __init__(self, fabricante, modelo):
        '''
        Cria uma representação de uma Televisão

        Parâmetros
        ----------
        fabricante:str
            Nome do fabricante da televisão
        modelo:str
            Modelo da Televisão
        
        Parâmetros Opcionais
        --------------------
        canal_atual:int
            Canal atual da televisão, inicializado em 0
        lista_canais:list
            Lista com canais sintonizados na televisão
        volume:int
            volume da televisão, inicializado em 0
        '''
        self.fabricante = fabricante
        self.modelo = modelo
        self.canal_atual = 0
        self.lista_canais = []
        self.volume = 0
        
class ControleRemoto:
    
    def __init__(self, televisao):
        '''
        Cria uma representação de um Controle Remoto

        Parâmetros
        ----------
        televisao:Televisao
            Recebe um Objeto da Classe Televisao
        self.televisao = televisao
        '''
    def ajustar_volume(self, volume):
        '''
        Ajusta o volume da televisão
        
        Parâmetros
        ----------
        volume:int
            Volume desejado para ajuste na televisão
        
        Retorno
        ----------
        out: Bool
            True se deu certo, False se não deu
        '''
        if volume >= 0 and volume <= 100:
            self.volume = volume
            return True
        return False
            
    def sintonizar_canal(self, canal_a_ser_sintonizado):
        '''
        Sintoniza um canal na televisão
        
        Parâmetros
        ----------
        canal_a_ser_sintonizado:int
            Canal desejado para ser sintonizado na televisão
        
        Retorno
        ----------
        out: Bool
            True se deu certo, False se não deu
        '''
        self.lista_canais = []
        if canal_a_ser_sintonizado not in self.lista_canais:
            self.lista_canais.append(canal_a_ser_sintonizado)
            return True
        return False 
        
    def trocar_canal(self, canal_desejado):
        '''
        Ajusta o volume da televisão

        Parâmetros
        ----------
        canal_desejado:int
            Canal desejado para ajuste na televisão

        Retorno
        ----------
        out: Bool
            True se deu certo, False se não deu
        '''
        if canal_desejado in self.lista_canais:
            self.canal_atual = canal_desejado
            return True
        return False
    
```

```python
tv1 = Televisor(
    fabricante = 'Samsung',
    modelo = 'J5500'
)
```

```python
#Testes
controle1 = ControleRemoto(tv1)
controle1.ajustar_volume(5)
print(controle1.volume)
controle1.sintonizar_canal(16)
print(controle1.lista_canais)
controle1.trocar_canal(16)
print(controle1.canal_atual)
```
